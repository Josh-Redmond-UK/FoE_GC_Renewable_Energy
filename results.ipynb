```python
import pandas as pd
```

```python
minres = pd.read_csv('min.csv')
medres = pd.read_csv('med.csv')
maxres = pd.read_csv('max.csv')
```

```python
minMed = pd.merge(minres, medres, on='area name', suffixes=(' min', ' med'))
```

```python
minMed = minMed[['area name', 'solar power area min', 'wind powe area min', 'solar power area med', 'wind powe area med']]
```

```python
full = pd.merge(minMed, maxres, on='area name', suffixes=(' ', ' max'))
```
